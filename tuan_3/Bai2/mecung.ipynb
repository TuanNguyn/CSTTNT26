# Bài 2: Mê cung (tìm đường đi cho con kiến)

Trong bài này ta biểu diễn mỗi ô của mê cung là một đỉnh của đồ thị, giống như trong file *my_lesson4.ipynb*:
- Mỗi ô trống là 1 đỉnh: ký hiệu bằng tọa độ `(r, c)` hoặc chuỗi `'r,c'`.
- Giữa hai ô kề nhau theo 4 hướng (lên, xuống, trái, phải) không có tường thì có 1 cạnh.
- Sau đó dùng thuật toán BFS để tìm đường đi ngắn nhất từ ô con kiến đến ô `A`.

```python
# ===== PHẦN 1: KHAI BÁO MÊ CUNG =====

# Kích thước mê cung (số hàng, số cột)
rows = 6   # số hàng
cols = 6   # số cột

# Danh sách các ô là TƯỜNG theo đề bài
# Mỗi ô được ghi bằng tọa độ (r, c) với r, c tính từ 0
# Ví dụ: tường ở hàng 2 cột 3 thì ghi (1, 2)
walls = [
    # (1, 2), (2, 2), ...
 ]

# Vị trí con kiến (ô xuất phát) và ô A (đích)
# Giả sử con kiến ở ô góc trên bên trái, A ở góc dưới bên phải
start_cell = (0, 0)
goal_cell  = (rows - 1, cols - 1)
```

```python
# ===== PHẦN 2: CHUYỂN MÊ CUNG THÀNH ĐỒ THỊ =====

def build_maze_graph(rows, cols, walls):
    graph = {}
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for r in range(rows):
        for c in range(cols):
            if (r, c) in walls:
                continue

            node = str(r) + "," + str(c)
            graph[node] = []

            for dr, dc in directions:
                nr = r + dr
                nc = c + dc
                if 0 <= nr < rows and 0 <= nc < cols and (nr, nc) not in walls:
                    neighbor = str(nr) + "," + str(nc)
                    graph[node].append(neighbor)

    return graph

maze_graph = build_maze_graph(rows, cols, walls)

print("Đồ thị mê cung (in vài đỉnh đầu):")
count = 0
for key in maze_graph:
    print(key, "->", maze_graph[key])
    count += 1
    if count >= 5:
        break
```

```python
# ===== PHẦN 3: THUẬT TOÁN BFS & HÀM TRUY VẾT =====

def BFS(graph, start, goal):
    """Tìm đường đi bằng BFS, trả về mảng truy vết (path)."""
    if start not in graph or goal not in graph:
        return None

    queue = []        # hàng đợi các đỉnh sẽ xét
    visited = []      # danh sách các đỉnh đã thăm
    path = {}         # path[x] = đỉnh đứng trước x trên đường đi

    # đưa đỉnh xuất phát vào hàng đợi
    queue.append(start)
    visited.append(start)
    path[start] = None

    while len(queue) > 0:
        current = queue.pop(0)  # lấy phần tử đầu hàng đợi

        # nếu gặp đích thì dừng
        if current == goal:
            break

        # duyệt các đỉnh kề với current
        for neighbor in graph[current]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.append(neighbor)
                path[neighbor] = current

    return path

def find_path(path, start, goal):
    """Từ mảng truy vết path, trả về danh sách các đỉnh đi từ start -> goal."""
    if path is None or goal not in path:
        return []

    result = []
    current = goal
    while current is not None:
        result.append(current)
        current = path[current]

    result.reverse()
    return result
```

```python
# ===== PHẦN 4: CHẠY THỬ TÌM ĐƯỜNG ĐI =====

# chuyển ô (r, c) thành chuỗi "r,c"
start_node = str(start_cell[0]) + "," + str(start_cell[1])
goal_node  = str(goal_cell[0])  + "," + str(goal_cell[1])

path_dict = BFS(maze_graph, start_node, goal_node)
path_nodes = find_path(path_dict, start_node, goal_node)

print("Đường đi từ", start_node, "đến", goal_node, ":")
print(path_nodes)
```
