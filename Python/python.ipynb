### Trò chơi đoán từ


```python
# # trò  chơi đoán từ 
secret_word = "python"
hint=" gợi ý đây là 1  ngôn ngữ lập trình giống với 1 con vật dài thòn khôn chân"
guess=""
guess_count = 0
guess_limit = 3
print(hint)

while guess!=secret_word:
    if guess_count < guess_limit:
        guess = input("Bạn đoán từ gì? ")
        guess_count +=1
        if guess != secret_word and guess_count < guess_limit:
            print("nhap lai !")
    else:
        break
    
if guess == secret_word:
    print("Chính xác ! ")
else: 
    print("Thua rồi !")
```

### Máy tính lũy thừa

```python
#máy tính tính lũy thừa
#exponent là số mũ
def calculate_power(number,exponent):
    result = 1
    for i in range(exponent):
        result *= number
    return result

print("Nhap number va so mũ:")
so =int(input("nhap so number: "))
mu = int(input("nhap so mũ: "))
print(calculate_power(so,mu))
```

# Thực hành với bài quizzes 📋

```python
# định nghĩa hàm quiz bằng class
class quiz:
    def __init__(self,question , answer):
        self.question = question
        self.answer = answer
        
```

```python
question = [
    "câu 1: ai là người góp phần lớn trong công cuộc giải phóng Việt Nam? \n a.Hồ Chí Minh \n b. Võ Nguyên Giáp \n c.Phan Châu Chinh",
    "câu 2: Thực thể nào đang góp phần thay đổi cuộc chơi của mỗi người trong Công Nghệ? \n a.dev \n b.Ai \n c.Con người", 
]

quizzes =[
    quiz(question[0],"a"),
    quiz(question[1],"b")
]

def run_quiz (quizzes):
    score = 0
    for Quiz in quizzes:
        print(Quiz.question)
        user_input = input("Nhập đáp án: ")
        if(user_input == Quiz.answer):
            score+=1
        print(f"Kết quả bạn trả lời được {score} / {len(quizzes)} ")
run_quiz(quizzes)
```
