# Phần bài tập thực hành 🗝️

## Bài 1: Liệt kê tổ hợp

```python
class ToHop:

    def __init__(self):
        self.n = 0
        self.k = 0
        self.elements = []
        self.result = []

    def input_data(self):
        self.n, self.k = map(int, input("Nhập N K: ").split())
        for i in range(self.n):
            self.elements.append(input(f"Phần tử {i+1}: ").strip())

    def backtrack(self, start, current):
        if len(current) == self.k:
            self.result.append(current[:])
            return
        for i in range(start, self.n):
            current.append(self.elements[i])
            self.backtrack(i + 1, current)
            current.pop()

    def solve(self):
        self.backtrack(0, [])
        return self.result

    def display_result(self):
        print(len(self.result))
        for combo in self.result:
            print(' '.join(combo))


# Test
tohop = ToHop()
tohop.input_data()
tohop.solve()
tohop.display_result()
```

## Bài 2: Xếp hậu

```python
class XepHau:

    def __init__(self):
        self.n = 0
        self.solutions = []
        self.col = set()
        self.diag1 = set()
        self.diag2 = set()

    def input_data(self):
        self.n = int(input("Nhập N: "))

    def is_safe(self, row, col):
        if col in self.col:
            return False
        if row - col in self.diag1:
            return False
        if row + col in self.diag2:
            return False
        return True

    def backtrack(self, row, current_solution):
        if row == self.n:
            self.solutions.append(current_solution[:])
            return
        for col in range(self.n):
            if self.is_safe(row, col):
                current_solution.append(col)
                self.col.add(col)
                self.diag1.add(row - col)
                self.diag2.add(row + col)
                self.backtrack(row + 1, current_solution)
                current_solution.pop()
                self.col.remove(col)
                self.diag1.remove(row - col)
                self.diag2.remove(row + col)

    def solve(self):
        self.backtrack(0, [])
        return self.solutions

    def display_result(self):
        if not self.solutions:
            print(0)
        else:
            print(1)
            solution = self.solutions[0]
            for row in range(self.n):
                line = []
                for col in range(self.n):
                    line.append('1' if col == solution[row] else '0')
                print(' '.join(line))


# Test
xephau = XepHau()
xephau.input_data()
xephau.solve()
xephau.display_result()
```

## Bài 3: Mã di tuần

```python
class MaDiTuan:

    def __init__(self):
        self.n = 0
        self.start_x = 0
        self.start_y = 0
        self.board = []
        self.moves = [(2, 1), (2, -1), (-2, 1), (-2, -1),
                      (1, 2), (1, -2), (-1, 2), (-1, -2)]
        self.solution = None

    def input_data(self):
        self.n, self.start_x, self.start_y = map(int, input("Nhập N x0 y0: ").split())
        self.board = [[-1] * self.n for _ in range(self.n)]

    def is_valid(self, x, y):
        return 0 <= x < self.n and 0 <= y < self.n and self.board[x][y] == -1

    def count_next(self, x, y):
        c = 0
        for dx, dy in self.moves:
            if self.is_valid(x + dx, y + dy):
                c += 1
        return c

    def solve(self):
        x, y = self.start_x, self.start_y
        self.board[x][y] = 0
        for step in range(1, self.n * self.n):
            nexts = []
            for dx, dy in self.moves:
                nx, ny = x + dx, y + dy
                if self.is_valid(nx, ny):
                    nexts.append((self.count_next(nx, ny), nx, ny))
            if not nexts:
                return None
            nexts.sort()
            _, x, y = nexts[0]
            self.board[x][y] = step
        self.solution = [row[:] for row in self.board]
        return self.solution

    def display_result(self):
        if not self.solution:
            print(0)
        else:
            print(1)
            for row in self.solution:
                print(' '.join(str(x) for x in row))


# Test
madituan = MaDiTuan()
madituan.input_data()
madituan.solve()
madituan.display_result()
```

## Bài 4: Tô màu đồ thị

```python
class ToMauDoThi:

    def __init__(self):
        self.vertices = 0
        self.adj = []
        self.colors = []
        self.num_colors = 0
        self.solution = None

    def input_data(self):
        self.vertices = int(input("Nhập số đỉnh n: "))
        self.adj = [[] for _ in range(self.vertices + 1)]
        m = int(input("Nhập số cạnh m: "))
        for _ in range(m):
            u, v = map(int, input("Cạnh u v: ").split())
            self.adj[u].append(v)
            self.adj[v].append(u)

    def is_safe(self, vertex, color):
        for neighbor in self.adj[vertex]:
            if self.colors[neighbor] == color:
                return False
        return True

    def backtrack(self, vertex):
        if vertex == self.vertices + 1:
            self.solution = self.colors[:]
            return True
        for color in range(1, self.num_colors + 1):
            if self.is_safe(vertex, color):
                self.colors[vertex] = color
                if self.backtrack(vertex + 1):
                    return True
                self.colors[vertex] = 0
        return False

    def solve(self):
        for k in range(1, self.vertices + 1):
            self.num_colors = k
            self.colors = [0] * (self.vertices + 1)
            if self.backtrack(1):
                return True
        return False

    def display_result(self):
        if not self.solution:
            print(0)
        else:
            print(self.num_colors)
            for color in range(1, self.num_colors + 1):
                group = [str(v) for v in range(1, self.vertices + 1) if self.solution[v] == color]
                print(' '.join(group))


# Test
tomau = ToMauDoThi()
tomau.input_data()
tomau.solve()
tomau.display_result()
```

# Phần bài tập về nhà 📋

## Bài 1: Liệt kê hoán vị

```python
class HoanVi:

    def __init__(self):
        self.n = 0
        self.elements = []
        self.result = []
        self.used = []

    def input_data(self):
        self.n = int(input("Nhập N: "))
        for i in range(self.n):
            self.elements.append(input(f"Phần tử {i+1}: ").strip())
        self.used = [False] * self.n

    def backtrack(self, current):
        if len(current) == self.n:
            self.result.append(current[:])
            return
        for i in range(self.n):
            if not self.used[i]:
                self.used[i] = True
                current.append(self.elements[i])
                self.backtrack(current)
                current.pop()
                self.used[i] = False

    def solve(self):
        self.backtrack([])
        return self.result

    def display_result(self):
        print(len(self.result))
        for perm in self.result:
            print(' '.join(perm))


# Test
hoanvi = HoanVi()
hoanvi.input_data()
hoanvi.solve()
hoanvi.display_result()
```

## Bài 2: Điền bảng số

```python
class DienBangSo:

    def __init__(self):
        self.board = [[0] * 9 for _ in range(9)]
        self.solution = None

    def input_data(self):
        for i in range(9):
            line = input(f"Hàng {i+1}: ").strip()
            for j in range(9):
                if line[j] == '.':
                    self.board[i][j] = 0
                else:
                    self.board[i][j] = int(line[j])

    def is_valid(self, row, col, num):
        if num in self.board[row]:
            return False
        for i in range(9):
            if self.board[i][col] == num:
                return False
        sr, sc = (row // 3) * 3, (col // 3) * 3
        for i in range(sr, sr + 3):
            for j in range(sc, sc + 3):
                if self.board[i][j] == num:
                    return False
        return True

    def find_empty(self):
        for i in range(9):
            for j in range(9):
                if self.board[i][j] == 0:
                    return (i, j)
        return None

    def backtrack(self):
        empty = self.find_empty()
        if empty is None:
            self.solution = [row[:] for row in self.board]
            return True
        row, col = empty
        for num in range(1, 10):
            if self.is_valid(row, col, num):
                self.board[row][col] = num
                if self.backtrack():
                    return True
                self.board[row][col] = 0
        return False

    def solve(self):
        self.backtrack()
        return self.solution

    def display_result(self):
        if not self.solution:
            print("IMPOSSIBLE")
        else:
            for row in self.solution:
                print(''.join(str(x) for x in row))


# Test
dienbangso = DienBangSo()
dienbangso.input_data()
dienbangso.solve()
dienbangso.display_result()
```
