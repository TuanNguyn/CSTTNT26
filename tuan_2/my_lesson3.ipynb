```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# Sử dụm hàm np.random.rand() sinh số ngẫu nhiên trong [0,1) --> [0,1) * (b - a) + a --> [a, b)
# Sử dụng hàm sau để tính hàm xác suất chuẩn N(m, s): np.sqrt, np.exp
mu =1 
sigma = 1.0
N=100

x_min = mu - 5 * sigma
x_max= mu + 5 * sigma

x = np.random.rand(N) * (x_max - x_min) + x_min

he_so = 1 / (sigma * np.sqrt(2 * np.pi))

phan_mu = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))
y = he_so * phan_mu

print("b. Sinh cặp (x, y) với y làm hàm phân phối chuẩn")
print(f"+ x: {x_min} <= min [{np.min(x): .2f}] <= max [{np.max(x): .2f}] <= {x_max}")
print(f"+ Ve ham so y(x)")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.scatter(x, y)              # ve cac diem (xi, yi)
plt.savefig("4b.png")          #  luu do thi
plt.show()                     #  hien thi do thi
```

#### 4. Giải hệ phương trình tuyến tính
Giải hệ phương trình tuyến tính sau:
$$
\left\{\begin{matrix}
        4x_1  + 3x_2 - 5x_3 = 2 \\ 
        -2x_1 - 4x_2 + 5x_3 = 5 \\ 
        8x_1 + 8x_2 = -3 \\ 
\end{matrix}\right.
$$

```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# Tạo ma trận A (hệ số)
A = np.array([
    [4,  3, -5],
    [-2, -4, 5],
    [8,  8,  0]
], dtype=float)

y = np.array([2, 5, -3], dtype=float)

x = np.linalg.solve(A, y)

yy = A @ x

diff = np.linalg.norm(yy - y)

print(f"+ Ma tran A: \n{A}")
print(f"+ Vector y: {y}")
print(f"+ Nghiem x: {x}")
print(f"+ Kiem tra: yy = Ax = {yy} ==> ||yy - y||_2 = {diff: .2f}")

print(f"+ Ma tran A: \n{A}")
print(f"+ Vector y: {y}")
print(f"+ Nghiem x: {x}")
print(f"+ Kiem tra: yy = Ax = {yy} ==> ||yy -y ||_2 = {diff: .2f}")
```

### Bài 3. Tính gần đúng số pi dùng phương pháp Monte Carlo

Vẽ một đường tròn $C$ có bán kính $r$ nội tiếp một hình vuông $C$, và đặt ngẫu nhiên $N$ chấm lên hình vuông. Tỉ lệ các chấm nằm trong hình tròn trên tổng số chấm xấp xỉ bằng diện tích của hình tròn chia cho hình vuông như sau:
$$
\frac{N_C}{N} \approx  \frac{\mathbb{S}_C}{\mathbb{S}_N}
$$
nơi mà $N_C$ là các chấm trong hình tròn, $N$ là tổng số chấm, $\mathbb{S}_C$ và $\mathbb{S}_N$ lần lượt là diện tích của hình tròn và hình vuông.
$$ \mathbb{S}_C = \pi * r^2 $$
$$ \mathbb{S}_N = (2*r)^2 = 4r^2 $$
Do đó, công thức trên trở thành:
$$
\frac{N_C}{N} \approx \frac{\pi * r^2}{4r^2} = \frac{\pi}{4}
$$
Suy ra, ta có giá trị số $\pi$ được tính như sau:
$$
\pi \approx \frac{4N_C}{N}
$$

Các bạn hãy lập trình tính sấp xỉ số $\pi$ dùng phương pháp Monte Carlo. Cho biết các sai số với $N = 100$, $N = 10000$, và $N = 1000000$. Vẽ đồ thị minh họa cho chương trình.

**Hướng dẫn**: Viết chương trình phát sinh $N$ điểm $(x_i, y_i)$ với $x_i, y_i \in [-1, -1]$. $(x_i, y_i)$ thuộc hình tròn tâm $(0, 0)$ bán kính 1 nội tiếp hình vuông có tâm tại $(0, 0)$ độ dài cạnh 2 khi và chỉ khi $x_i^2 + y_i^2 <= 1$.

```python
import matplotlib.pyplot as plt # khai báo thư viện vẽ pyplot
import math

circle = plt.Circle((0, 0), 1, color='g') # tạo đối tượng vòng tròn

fig, ax = plt.subplots(figsize=(6, 6)) # lấy figure và vùng vẽ

plt.plot(0, 0, 'o', color=(0.9, 0.9, 1.0), alpha=0.8) # vẽ điểm tại tâm (0,0)
ax.add_patch(circle) # thêm vòng tròn vào vùng vẽ

x = 0.75; y = math.sqrt(1 - (x ** 2)) # (x = 0.75, y = căn bậc 2 của 1 - x^2)
plt.arrow(0, 0, x, y)

plt.xlim(-1, 1) # giới hạn trục x
plt.ylim(-1, 1) # giới hạn trục y

plt.savefig('pi.png') # lưu đồ thị xuống tập tin
plt.show()            # hiển thị đồ thị
```

```python
import numpy as np

def calc_pi_monte_carlo(n = 100):

    x = np.random.rand(n) * 2 - 1  # [0, 1) -> [-1, 1)
    y = np.random.rand(n) * 2 - 1
 
    distance_squared = x**2 + y**2

    inside_circle = distance_squared <= 1
    n_c = np.sum(inside_circle)  # Số điểm trong hình tròn
 
    pi = 4 * n_c / n
    
    return pi

# calc_pi_monte_carlo
```

```python
import math

print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
```
